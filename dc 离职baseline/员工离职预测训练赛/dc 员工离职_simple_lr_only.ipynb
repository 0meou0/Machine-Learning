```python
import pandas as pd
pd.set_option('display.max_columns', None)
path='./'
train = pd.read_csv(path+'input/pfm_train.csv')
test = pd.read_csv(path+'input/pfm_test.csv')
```

```python
test['Attrition']=-1
data = train.append(test).reset_index(drop=True)
```

```python
data.drop(['Over18', 'StandardHours'], axis=1, inplace=True)
```

```python
cat_col = [i for i in data.select_dtypes(object).columns if i not in ['Attrition']]
```

```python
def encode_onehot(df,column_name):
    feature_df=pd.get_dummies(df[column_name], prefix=column_name)
    all = pd.concat([df.drop([column_name], axis=1),feature_df], axis=1)
    return all
```

```python
for i in cat_col:
    data = encode_onehot(data,i)
```

```python
data.drop(['EmployeeNumber'], axis=1, inplace=True)
# 删除唯一值
```

```python
data.head()
```

```python
feats = [i for i in data.columns if i not in ['Attrition']]
```

```python
import numpy as np

max_min_scaler = lambda x : (x-np.min(x))/(np.max(x)-np.min(x))
for i in feats:
    data[i]=data[[i]].apply(max_min_scaler)
```

```python
data.head()
```

```python
data.corr()
```

```python
# 画相关性热力图
import matplotlib.pyplot as plt
import seaborn as sns
def correlation_heatmap(data):
    correlations = data.corr()
    fig, ax = plt.subplots(figsize=(30,30))
    sns.heatmap(correlations, vmax=1.0, center=0, fmt='.2f',
                square=True, linewidths=.5, annot=True, cbar_kws={"shrink": .70})
    plt.show();
correlation_heatmap(data)
```

```python
X=np.array(data[feats])
target = data['Attrition']
```

```python
X
```

```python
X.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_std=sc.fit_transform(X)
```

```python
X_std
```

```python
from sklearn import decomposition 
pca=decomposition.PCA(n_components=20)
#使用PCA对特征进行降维

X_std_pca=pca.fit_transform(X_std)
#查看降维后特征的维度
data = pd.DataFrame(X_std_pca)
data['Attrition'] = target
X_std_pca.shape
```

```python
data
```

```python
feats = [i for i in data.columns if i not in ['Attrition']]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
train = data[data['Attrition'] !=-1][feats]
train=train.fillna(0)
label = data[data['Attrition'] !=-1]['Attrition']
test = data[data['Attrition'] ==-1][feats]
test = test.fillna(0)
model = LogisticRegression(max_iter=1000)
model.fit(train,label)
```

#### 直接得到单模结果

```python
test.shape
```

```python
p =model.predict_proba(test)[:,1]
test['result']=p
test[['result']].to_csv(path+'output/lr_only_proba.csv',index=False)
```

```python
del test['result']
z =model.predict(test)
test['result']=z
test[['result']].to_csv(path+'output/lr_only_submission.csv',index=False)
```

使用网络搜索选择最优参数

```python
from sklearn.model_selection import GridSearchCV
lg=LogisticRegression()
clf=GridSearchCV(lg,param_grid=[{'C':np.arange(0.001,0.05,0.001)}],cv=5)
clf.fit(train,label)
best_model=clf.best_estimator_
clf.best_params_
```

```python
del test['result']
result_gs=best_model.predict(test)
# pd.DataFrame({'result':result_gs}).to_csv(path+'output/grid_lrsubmission.csv',index=False)
```
